```python
from datasets import load_dataset, load_from_disk, Dataset
import numpy as np
import pandas as pd
from transformers import RobertaTokenizer, RobertaModel
import torch
from torch.utils.data import TensorDataset, DataLoader
```

```python
data_base = 'data'
# data_base = ''

dataset_name = 'cnn_dailymail'

data_path = data_base+'/'+dataset_name
```

```python
dataset = load_from_disk(
    'sentence_embs_'
)

val_sentence_data = dataset['validation']
```

```python
dataset
```

```python
val_df = pd.DataFrame(dataset['validation'])
```

```python
batch_size = 40

val_embeddings = []
for i in range(0, val_sentence_data.num_rows, batch_size):
    val_embeddings.append(
        val_df.iloc[i:i+batch_size]['embeddings'].tolist()
    )
    if i//batch_size%100==0:
        print('Loaded',i//batch_size,'batches of',val_sentence_data.num_rows//batch_size,'batches')
```

```python
len(val_embeddings)
```

```python
import pickle

part = 1
start = 20000*(part-1)
end = 20000*(part)
with open('val_part_'+str(part)+'.pkl', 'wb') as pickle_file:
    pickle.dump(val_embeddings[start:end], pickle_file, protocol=pickle.HIGHEST_PROTOCOL)
```
